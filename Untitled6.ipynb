```python
Questions 1
-----------------------------------------------
def twoSum(nums, target):
    # Sort the array in ascending order
    nums.sort()

    # Initialize pointers
    left = 0
    right = len(nums) - 1

    # Iterate while left < right
    while left < right:
        # Calculate the sum
        current_sum = nums[left] + nums[right]

        if current_sum == target:
            # Found the pair that adds up to the target
            return [left, right]
        elif current_sum < target:
            # Increment the left pointer
            left += 1
        else:
            # Decrement the right pointer
            right -= 1

    # No two numbers add up to the target
    return []

# Example usage
nums = [2, 7, 11, 15]
target = 9
result = twoSum(nums, target)
print(result)  # Output: [0, 1]
```

```python
Questions 2
----------------------------------------------------------
def removeElement(nums, val):
    i = 0  # Pointer to track the elements not equal to val

    for j in range(len(nums)):
        if nums[j] != val:
            nums[i] = nums[j]
            i += 1

    return i

# Example usage
nums = [3, 2, 2, 3]
val = 3
result = removeElement(nums, val)
print("Count:", result)  # Output: 2
print("Updated nums:", nums[:result])  # Output: [2, 2]
```

```python
Questions 3
---------------------------------------------------------
def searchInsert(nums, target):
    left = 0
    right = len(nums) - 1

    while left <= right:
        mid = (left + right) // 2

        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return left

# Example usage
nums = [1, 3, 5, 6]
target = 5
result = searchInsert(nums, target)
print(result)  # Output: 2
```

```python
Questions 5
-----------------------------------------------------------------
def merge(nums1, m, nums2, n):
    p1 = m - 1  # Pointer for nums1
    p2 = n - 1  # Pointer for nums2
    p = m + n - 1  # Pointer for merged array

    while p1 >= 0 and p2 >= 0:
        if nums1[p1] >= nums2[p2]:
            nums1[p] = nums1[p1]
            p1 -= 1
        else:
            nums1[p] = nums2[p2]
            p2 -= 1
        p -= 1

    # Copy remaining elements from nums2 to nums1
    nums1[:p2 + 1] = nums2[:p2 + 1]

# Example usage
nums1 = [1, 2, 3, 0, 0, 0]
m = 3
nums2 = [2, 5, 6]
n = 3
merge(nums1, m, nums2, n)
print(nums1)  # Output: [1, 2, 2, 3, 5, 6]
```

```python
Questions 6
-----------------------------------------------------------------
def containsDuplicate(nums):
    num_set = set()

    for num in nums:
        if num in num_set:
            return True
        else:
            num_set.add(num)

    return False

# Example usage
nums = [1, 2, 3, 1]
result = containsDuplicate(nums)
print(result)  # Output: True
```

```python
Questions 5
-----------------------------------------------------------------
def moveZeroes(nums):
    left = 0  # Pointer for non-zero elements

    for right in range(len(nums)):
        if nums[right] != 0:
            # Swap non-zero element to the left side
            nums[left], nums[right] = nums[right], nums[left]
            left += 1

# Example usage
nums = [0, 1, 0, 3, 12]
moveZeroes(nums)
print(nums)  # Output: [1, 3, 12, 0, 0]
```

```python
def findErrorNums(nums):
    xor_result = 0
    for i, num in enumerate(nums):
        xor_result ^= num
        xor_result ^= (i+1)

    # Find the rightmost set bit in xor_result
    rightmost_set_bit = xor_result & ~(xor_result - 1)

    missing = 0
    duplicate = 0

    for i, num in enumerate(nums):
        if num & rightmost_set_bit:
            missing ^= num
        if (i+1) & rightmost_set_bit:
            missing ^= (i+1)

    for num in nums:
        if num == duplicate:
            duplicate = num
        if num == missing:
            missing = num

    return [duplicate, missing]

# Example usage
nums = [1, 2, 2, 4]
result = findErrorNums(nums)
print(result)  # Output: [2, 3]
```
